```python
Q71="7.1. Write a python program to implement insertion sort using lists."
print(Q71)
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
# Example usage:
nums = [2, -8, 4, 5, 7]
insertion_sort(nums)
print("Insertion sort result:", nums)
```
