```python
Q2="""1.Calculation of Test Average: Write a python program to find the best of two test average marks out of three test’s marks"""
print(Q2)
input_str = input("Enter a string: ")
if input_str == input_str[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
for ch in set(input_str):
    # The line 't i t t t( h)' was a syntax error and has been removed.
    count = input_str.count(ch)
    print(f"'{ch}' appears {count} times")
```
